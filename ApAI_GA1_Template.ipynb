```python
from statistics_calculator import compute_statistics


def display_statistics(filename):
      """
      Wrapper function that computes and displays statistics for a CSV file.

      Args:
          filename (str): Path to the CSV file

      Returns:
          dict: Dictionary with column names as keys and statistics as values
      """
      data_structure_statistics = compute_statistics(filename)

      # Display the results
      for column_name, stats in data_structure_statistics.items():
          print(f'{column_name}:')
          print(f'  min: {stats["min"]:.2f}')
          print(f'  max: {stats["max"]:.2f}')
          print(f'  mean: {stats["mean"]:.2f}')
          print(f'  std_dev: {stats["std_dev"]:.2f}')
          print()

      return data_structure_statistics

import matplotlib.pyplot as plt

def plot_time_series(filename, title=None):
      """
      Plot time series data from a CSV file.

      Args:
          filename (str): Path to the CSV file
          title (str): Optional title for the plot
      """
      data = csv_parser(filename)

      # Get DateTime column if it exists
      datetime_col = None
      for col_name in data.keys():
          if 'DateTime' in col_name or 'Date' in col_name:
              datetime_col = col_name
              break

      if datetime_col is None:
          print("No DateTime column found")
          return

      # Plot numeric columns
      numeric_cols = [col for col in data.keys() if col != datetime_col]

      if not numeric_cols:
          print("No numeric columns to plot")
          return

      # Create subplots
      fig, axes = plt.subplots(len(numeric_cols), 1, figsize=(10, 3*len(numeric_cols)))

      if len(numeric_cols) == 1:
          axes = [axes]

      for idx, col_name in enumerate(numeric_cols):
          axes[idx].plot(data[datetime_col], data[col_name], linewidth=0.8)
          axes[idx].set_ylabel(col_name)
          axes[idx].grid(True, alpha=0.3)
          axes[idx].set_title(f'{col_name} over time')

      axes[-1].set_xlabel('Date')

      if title:
          fig.suptitle(title, fontsize=14, y=1.00)

      plt.tight_layout()
      plt.show()
```

**Sensor drift and failure** accounts for ~40% of data quality issues in automated weather stations
**Data entry errors** occur during quality control when technicians manually correct suspicious values
**Transmission errors** happen due to network issues, especially in remote stations

## Analysis: Can Basic Statistics Identify the Incorrect Data?

  ### Summary of Changes

  The introduction of four data errors had the following impact on the **Temperature** column statistics:

  | Statistic | Original | Modified | Change | % Change |
  |-----------|----------|----------|--------|----------|
  | **Minimum** | -1.81°C | **-28.90°C** | -27.09°C | **+1497%** |
  | **Maximum** | 26.38°C | **48.20°C** | +21.82°C | **+83%** |
  | **Mean** | 11.14°C | 11.25°C | +0.11°C | +1% |
  | **Std Dev** | 5.35 | 6.29 | +0.94 | **+18%** |

  ### Are Basic Statistics Sufficient?

  **Partial Success:** Basic statistics (min/max/mean/std_dev) provide **mixed results** in identifying data quality issues:

  #### What Statistics Successfully Detected:

  1. **Min/Max Changes:**
     The dramatic changes in minimum (-1497%) and maximum (+83%) values clearly indicate outliers
     These extreme values are **easily detectable** and flag potential sensor malfunctions

  2. **Standard Deviation Increase (+18%):**
     The 18% increase in std_dev signals increased data variability
     Suggests presence of outliers, though doesn't pinpoint their location

  #### What Statistics Failed to Detect:

  1. **Mean Barely Changed (+1%):**
     With only 4 errors in 356 data points (~1%), the mean is **resistant to outliers**
     The mean of 11.25°C appears normal, masking the data quality issues

  2. **No Location Information:**
     Statistics don't reveal **where** or **when** errors occurred
     Cannot distinguish between: (a) one extreme outlier, (b) multiple moderate outliers, (c) systematic drift

  3. **Logical Inconsistencies Undetected:**
     The error where temp_low (15.4°C) > temp (8.56°C) is **completely invisible** in summary statistics
     Basic statistics don't capture inter-column relationships


  To properly identify and locate data quality issues, we need to do the following:

  Identify outliers using IQR method and Rolling Statistics (Moving Windows)
  and Time Series Visualization, Plotting data over time immediately reveals the four anomalous spikes, Visual inspection is far more effective than summary statistics alone
